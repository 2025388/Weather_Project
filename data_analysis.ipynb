```python
import pymysql
import pandas as pd

# MySQL database connection configuration
db_config = {
    'host': 'localhost',
    'port': 3306,
    'user': 'root',
    'password': 'Hamilton1186!',
    'database': 'era5_data'
}

def fetch_first_20_rows():
    try:
        # Connect to the MySQL database
        conn = pymysql.connect(**db_config)
        cursor = conn.cursor()

        # Query to fetch the first 20 rows from era5_summary
        query = "SELECT * FROM era5_summary LIMIT 20"
        cursor.execute(query)

        # Fetch column names
        columns = [col[0] for col in cursor.description]

        # Fetch all rows
        rows = cursor.fetchall()

        if not rows:
            print("No data found in the era5_summary table.")
            return

        # Create a DataFrame for better visualization
        df = pd.DataFrame(rows, columns=columns)

        # Close cursor and connection
        cursor.close()
        conn.close()

        # Print the DataFrame
        print("\nFirst 20 rows from era5_summary table:")
        print(df.to_string(index=False))

    except pymysql.Error as e:
        print(f"Error connecting to MySQL or fetching data: {e}")

if __name__ == "__main__":
    fetch_first_20_rows()
```
